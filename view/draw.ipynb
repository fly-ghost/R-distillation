# 绘制相关图标，标签使用8 Times New Roman

```python
import json
import os

import matplotlib.pyplot as plt
```

```python
def draw_variation(filename):
    path = os.path.join("codet5p-small", "feedback", filename)
    """
    T的影响
    """
    epochs = []
    T_list = []
    Q_list = []
    S_list = []
    C_list = []
    with open(path, "r") as f:
        for line in f:
            json_obj = json.loads(line)
            epochs.append(int(json_obj["id"]))
            T_list.append(json_obj["T"])
            Q_list.append(json_obj["Q"])
            S_list.append(json_obj["Success Rate"] * 100)
            C_list.append(json_obj["Compile Rate"] * 100)

    plt.figure(figsize=(12, 8))
    plt.subplot(2, 2, 1)
    plt.plot(epochs, T_list)
    plt.xticks(epochs)
    plt.xlabel("epochs")
    plt.ylabel("Temperatrue")

    plt.subplot(2, 2, 2)
    plt.plot(epochs, Q_list, c="green")
    plt.xticks(epochs)
    plt.xlabel("epochs")
    plt.ylabel("Q")

    plt.subplot(2, 2, 3)
    plt.plot(epochs, S_list, c="purple")
    plt.xticks(epochs)
    plt.xlabel("epochs")
    plt.ylabel("Success Rate (%)")

    plt.subplot(2, 2, 4)
    plt.plot(epochs, C_list, c="orange")
    plt.xticks(epochs)
    plt.xlabel("epochs")
    plt.ylabel("Compile Rate (%)")

    plt.suptitle("The effect of T on distillation for codet5p-220m", fontdict={'family' : 'Times New Roman', 'size': 16})
    plt.show()
```

```python
draw_variation("values.jsonl")
```
